```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
h = [(10**(-k)) for k in range(0, 16)]
print(h)
```

```python
def function(x):
    return (np.tan(x))
```

```python
def derrivative(h, x):
    return (function(x+h) - function(x))/h
```

```python
actual_value = 1 + (np.tan(1)*np.tan(1))
print(actual_value)

arrayOfValues = [derrivative(h_value, 1) for h_value in h]
print(arrayOfValues)
```

```python
results = [ np.abs(arrayOfValues[i] - actual_value) for i in range(0, 16)]
print(results)
```

```python
fig, plot = plt.subplots()
plot.loglog(h, results)

plot.set_xlabel("h value")
plot.set_ylabel("Error Value")
```

```python
def derrivative2(h, x):
    return (function(x+h) - function(x-h))/(2*h)
```

```python
result2 = [np.abs(derrivative2(h_value, 1) - actual_value) for h_value in h]
```

```python
fig, plot = plt.subplots()
plot.loglog(h, result2)

plot.set_xlabel("h value")
plot.set_ylabel("Error Value")
```

```python
epss = np.finfo(float).eps

minValue = min(results)
minValue2 = min(result2)

print(minValue, np.sqrt(epss))
print(minValue2, np.sqrt(epss))
```
